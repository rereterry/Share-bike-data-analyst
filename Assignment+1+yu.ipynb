```python
# part A-II-2
import pandas as pd
import numpy as np
import matplotlib

from matplotlib import pyplot as plt
import seaborn as sns
import os.path
userhome = os.path.expanduser('~')
csvfile = os.path.join(userhome, 'Desktop', 'Bike-Sharing-dataset', 'hour.csv')
df = pd.read_csv(csvfile)
df.describe()
```

```python
#part A-II-3 many outliers
import seaborn as sns
plt.figure(1)
sns.boxplot(x = 'hr', y='casual', data = df)
plt.figure(2)
sns.boxplot(x = 'hr', y='registered', data = df)
plt.figure(3)
sns.boxplot(x = 'weekday', y='registered', data = df)
plt.show()
```

```python
# information from this csv -- part A-II-1  and part A-III-1
plt.figure(4)
x =(df.hr)
y =(df.cnt)
plt.title('every hour user number')
plt.xlabel('hour')
plt.ylabel('people')
plt.bar(x,y)
plt.figure(5)
m =(df.temp)
plt.title("temperature influence")
plt.xlabel('Temperature')
plt.ylabel('people')
plt.scatter(m,y)
n = (df.mnth)
plt.figure(6)
plt.title("month v.s people")
plt.xlabel('Month')
plt.ylabel('people')
plt.bar(n,y)
plt.show()
```

```python
#part A-III-2 find every day the situation of riders and what period of times is users like
plt.figure(7)
sns.lmplot(x = 'hr', y='casual', hue='weekday', data = df)
plt.figure(8)
sns.lmplot(x = 'hr', y='registered', hue="weekday", data = df)
plt.show()
```
